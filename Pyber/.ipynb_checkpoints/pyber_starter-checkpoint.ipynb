```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
# Read the City and Ride Data
dfCity = pd.read_csv(city_data_to_load)
dfRide = pd.read_csv(ride_data_to_load)
```

```python
# Combine the data into a single dataset
dfMergePyber = pd.merge( dfRide,dfCity,on="city")
# Display the data table for preview
dfMergePyber.head()
```

```python
dfPyber_City = dfMergePyber.groupby(['city',"type"])
dfPyberCity_Fare = pd.DataFrame(dfPyber_City['fare'].mean()).reset_index()
dfPyberCity_Fare_Sum = pd.DataFrame(dfPyber_City['fare'].sum()).reset_index()
dfPyberCity_Drivers = pd.DataFrame(dfPyber_City['driver_count'].mean()).reset_index()
dfPyberCity_Rides = pd.DataFrame(dfPyber_City['ride_id'].count()).reset_index()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# urban
ridesUrban = dfPyberCity_Rides.loc[(dfPyberCity_Rides['type'] == "Urban"),'ride_id']
faresUrban = dfPyberCity_Fare.loc[(dfPyberCity_Fare['type'] == "Urban"),'fare']
driversUrban = dfPyberCity_Drivers.loc[(dfPyberCity_Drivers['type'] == "Urban"),'driver_count']
# suburban
ridesSubUrban = dfPyberCity_Rides.loc[(dfPyberCity_Rides['type'] == "Suburban"),'ride_id']
faresSubUrban = dfPyberCity_Fare.loc[(dfPyberCity_Fare['type'] == "Suburban"),'fare']
driversSubUrban = dfPyberCity_Drivers.loc[(dfPyberCity_Drivers['type'] == "Suburban"),'driver_count']
# rural
ridesRural = dfPyberCity_Rides.loc[(dfPyberCity_Rides['type'] == "Rural"),'ride_id']
faresRural = dfPyberCity_Fare.loc[(dfPyberCity_Fare['type'] == "Rural"),'fare']
driversRural = dfPyberCity_Drivers.loc[(dfPyberCity_Drivers['type'] == "Rural"),'driver_count']
```

```python
# Build the scatter plots for each city types
plt.scatter(ridesUrban,faresUrban, s=driversUrban*10,alpha=".8", facecolors="lightcoral", edgecolors="black",label="Urban")
plt.scatter(ridesSubUrban,faresSubUrban, s=driversSubUrban*10,alpha=".8", facecolors="lightskyblue", edgecolors="black",label="Suburban")
plt.scatter(ridesRural,faresRural, s=driversRural*10,alpha=".8", facecolors="gold", edgecolors="black",label="Rural")
# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()
# Create a legend
lgnd = plt.legend(title="City Types", loc="best",fancybox=True, framealpha=1)
lgnd.legendHandles[0]._sizes = [45]
lgnd.legendHandles[1]._sizes = [45]
lgnd.legendHandles[2]._sizes = [45]
# Incorporate a text label regarding circle size
plt.text(45, 30.5, 'Note:\nCircle size correlates with driver count per city.')
# Save Figure
plt.savefig("Pyber-Bubbleplot.png")
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
faresUrban_Total = dfPyberCity_Fare_Sum.loc[(dfPyberCity_Fare['type'] == "Urban"),'fare']
faresSubUrban_Total = dfPyberCity_Fare_Sum.loc[(dfPyberCity_Fare['type'] == "Suburban"),'fare']
faresRural_Total = dfPyberCity_Fare_Sum.loc[(dfPyberCity_Fare['type'] == "Rural"),'fare']
totalFares = faresUrban_Total.sum() + faresSubUrban_Total.sum() + faresRural_Total.sum()
perFaresUrban = faresUrban_Total.sum() / totalFares 
perFaresSuburban = faresSubUrban_Total.sum() / totalFares 
perFaresRural = faresRural_Total.sum() / totalFares 
# Build Pie Chart
labels = ["Urban", "Rural", "Suburban"]
sizes = [perFaresUrban,perFaresRural,perFaresSuburban]
colors = ["lightcoral","gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.title("% of Total Fares by City Type")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=-90)
# Save Figure
plt.savefig("Pyber-fares-by-city.png")
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
totalRides = ridesUrban.sum() + ridesSubUrban.sum() + ridesRural.sum()
perRidesUrban = ridesUrban.sum() / totalRides
perRidesSuburban = ridesSubUrban.sum() / totalRides
perRidesRural = ridesRural.sum() / totalRides 
# Build Pie Chart
labels = ["Urban", "Rural", "Suburban"]
sizes = [perRidesUrban,perRidesRural,perRidesSuburban]
colors = ["lightcoral","gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.title("% of Total Rides by City Type")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=-90)
# Save Figure
plt.savefig("Pyber-rides-by-city.png")
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
totalDrivers = driversUrban.sum() + driversSubUrban.sum() + driversRural.sum()
perDriversUrban = driversUrban.sum() / totalDrivers
perDriversSuburban = driversSubUrban.sum() / totalDrivers
perDriversRural = driversRural.sum() / totalDrivers 
# Build Pie Charts
labels = ["Urban", "Rural", "Suburban"]
sizes = [perDriversUrban,perDriversRural,perDriversSuburban]
colors = ["lightcoral","gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.title("% of Total Drivers by City Type")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=-120)
# Save Figure
plt.savefig("Pyber-drivers-by-city.png")
# Show Figure
plt.show()
```
